## Tensor DATATYPES

```python
dt = torch.tensor([1.0,2.0,3.0],
                  dtype=torch.complex32,
                  requires_grad=False)
dt
```

```python
dt.dtype
```

```python
dts = dt.type(torch.float32)
dts, dts.dtype
```

```python
dts * dt
```

```python
dts.shape
```

```python
print(f"DataType of dts and dt: {dts.dtype} and {dt.dtype}")
print(f"Shape of dts and dt: {dts.shape} and {dt.shape}")
print(f"Device of dts and dt: {dts.device} and {dt.device}")
```
